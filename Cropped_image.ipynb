```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

```

```python
img1 = cv2.imread('/content/drive/MyDrive/Assignment_Productize_Tech/1.jpg')
img2 = cv2.imread('/content/drive/MyDrive/Assignment_Productize_Tech/2.jpg')

img1 = cv2.resize(img1,(1000,600))
img2 = cv2.resize(img2,(1000,600))
```

```python
cv2_imshow(img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2_imshow(img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Select ROI
r1 = cv2.selectROI("select the area", img1)
# Crop image
cropped_image1 = img1[int(r1[1]):int(r1[1]+r1[3]),
                      int(r1[0]):int(r1[0]+r1[2])]
 
# Display cropped image
cv2.imshow("Cropped image1", cropped_image1)
#save cropped image
cv2.imwrite("cropped_image1.jpeg",cropped_image1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Select ROI
r2 = cv2.selectROI("select the area", img2)
# Crop image
cropped_image2 = img2[int(r2[1]):int(r2[1]+r2[3]),
                      int(r2[0]):int(r2[0]+r2[2])]
 
# Display cropped image
cv2.imshow("Cropped image2", cropped_image2)
#save cropped image
cv2.imwrite("cropped_image2.jpeg",cropped_image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def get_x_and_y(event):
    global lasx, lasy
    lasx, lasy = event.x, event.y

def draw_smth(event):
    global lasx, lasy
    canvas.create_line((lasx, lasy, event.x, event.y), 
                      fill='lightgreen', 
                      width=8)
    lasx, lasy = event.x, event.y
```

```python
from tkinter import *
app = Tk()
app.geometry("810x560")
canvas = Canvas(app, bg='black')
canvas.pack(anchor='nw', fill='both', expand=1)
from tkinter import *
canvas.bind("<Button-1>", get_x_and_y)
canvas.bind("<B1-Motion>", draw_smth)
image = PIL.Image.open("/content/drive/MyDrive/Assignment_Productize_Tech/cropped_image1.jpeg")
image = ImageTk.PhotoImage(image)
canvas.create_image(0,0, image=image, anchor='nw')

app.mainloop()
```
